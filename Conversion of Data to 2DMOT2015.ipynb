```python
import pandas as pd
import numpy as np
import matplotlib
import os
import subprocess
import csv
```

```python
path='C:/Users/asd/Desktop/500Frames'
```

```python
for i in range(1,17):
    inpath=path+"/"+str(i)
    for j in range(1,8):
        try:
            fullPath=inpath+"/"+str(j)+"/00"+str(j)+".csv"#j loop represents the name of the video folder and the video we are trying to extract
            readFile=pd.read_csv(fullPath)
            keep_col = ['frameNumber','id','x_TL','y_TL','x_BR','y_BR', 'isOccluded']
            newFile = readFile[keep_col]
            newFile['x_BR']=abs(newFile['x_BR']-newFile['x_TL'])###WIDTH
            newFile['y_BR']=abs(newFile['y_TL']-newFile['y_BR'])####HEIGHT
            for i in range(len(readFile)):
                if(newFile.isOccluded[i]==1):
                    newFile.isOccluded[i]=0#Confidence level is 0 when pedestrian is occluded in ground truth
                else:
                    newFile.isOccluded[i]=1#Confidence Level is 1 when pedestrian is not occluded in ground truth
            newFile.to_csv(inpath+"/"+str(j)+"/MOT"+str(j)+".csv", index=False)
        except:
            print("File Not Found",i,j)
```
